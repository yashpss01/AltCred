# Exploratory Data Analysis (EDA) - AltCred
This notebook contains the initial analysis of the credit scoring dataset.

## 1. Dataset Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../../data/raw/credit_score_dataset.csv') 
df.head()
```

## 2. Data Inspection

```python
df.info()
df.describe()
```

## 3. Missing Value Analysis

```python
df.isnull().sum()
```

## 4. Class Distribution

```python
if 'Credit_Score' in df.columns:
    sns.countplot(x='Credit_Score', data=df)
    plt.show()
```

## 5. Correlation Analysis

```python
plt.figure(figsize=(12,10))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
